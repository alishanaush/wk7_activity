# 2. Check whether female and male students scored the same marks

```python
#Null Hypothesis H0:female and male students scored the same marks
#Alt hypothesis Ha: female and male students didnt scored the same marks
#alpha=.05
```

```python
from scipy.stats import ttest_ind
```

```python
m=data[data['Gender']=='male']['Total Marks']
f=data[data['Gender']=='female']['Total Marks']
alpha=.05
```

```python
_,p = ttest_ind(m,f)
print("p value :", p)
if p < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0.")
```

Acc to the hypothesis,Female and male students scored the same marks

# 3.Whether test preparation helps the students?

```python
#Null Hypothesis H0:test preparation doesnt help the students
#Alt hypothesis Ha: test preparation help the students
```

```python
T_none=data[data['Test preparation']=='none']['Total Marks']
T_complt=data[data['Test preparation']=='completed']['Total Marks']
```

```python
_,p = ttest_ind(T_none,T_complt)
print("p value :", p)
if p < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0.")
```

```python
Acc to the hypothesis test preparation help students
```
